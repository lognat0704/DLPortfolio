```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [16, 9]
```

```python
org = cv2.cvtColor(cv2.imread('../facefactory/temp/test.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(org)
org.shape
```

```python
im = org.copy()
mask = np.zeros_like(im)
mask[60:230, 500:660, :] = 255 ## height, width, channel
im[60:230, 500:660, :] = 255
plt.imshow(im)
```

```python
#plt.imshow(im)
cv2.imwrite('examples/pic4.jpg', cv2.cvtColor(im, cv2.COLOR_RGB2BGR))
cv2.imwrite('examples/mask4.jpg', mask)
```
